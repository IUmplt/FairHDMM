```python
plt.rcParams['figure.figsize'] = 15, 15
plt.rcParams['axes.labelsize']=92
plt.rcParams['xtick.labelsize']=72
plt.rcParams['ytick.labelsize']=72
plt.rcParams['legend.loc']='lower right'
plt.rcParams['legend.fontsize'] = 48
plt.rcParams['lines.linewidth'] = 4
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 1
plt.rcParams['grid.color'] = 'k'
plt.rcParams['boxplot.boxprops.linewidth'] = 6
plt.rcParams['boxplot.whiskerprops.linewidth']=6
plt.rcParams['boxplot.flierprops.markeredgewidth']=4
plt.rcParams['boxplot.capprops.linewidth'] =6
plt.rcParams['boxplot.medianprops.color']='C2'
plt.rcParams['boxplot.medianprops.linewidth'] =6
plt.rcParams['boxplot.meanprops.markersize'] = 20
```

```python
modes = ['uni', 'fsum', 'buc_qsd']
mode_names = ['Util','WUtil', 'Water']
pos = np.array([ 1, 3, 5 ], dtype=int)
```

```python
data_path = './data/identity_total_new_n16_k10_s0.pkl'
fo = open(data_path, 'rb')
data = pickle.load(fo)
fo.close()
```

```python
total_errors = data['total_errors']
max_ratio_errors = data['max_ratio_errors']
inters = data['inters']
np.array(list(total_errors.values()))[pos]
```

```python
plt.bar(mode_names, np.array(list(total_errors.values()))[pos])
plt.grid()
#plt.ylabel('Total Errors')
#plt.tight_layout()
plt.savefig('adv_total_errors_k10.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.bar(mode_names, np.array(list(max_ratio_errors.values()))[pos])
#plt.ylabel('Max Ratio Errors')
#plt.ylim((0,1.1))
plt.grid()
plt.axhline(y=1, color='C3', linewidth=6)
plt.tight_layout()
plt.savefig('adv_max_ratio_errors_k10.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.bar(mode_names, np.array(list(inters.values()))[pos]+1)
plt.axhline(y=1, color='C3', linewidth=6)
plt.grid()
#plt.ylabel('Interferences')
plt.tight_layout()
plt.savefig('adv_inters_k10.pdf', bbox_inches='tight')
plt.show()
```
