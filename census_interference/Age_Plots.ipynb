```python
%matplotlib inline
import math
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import pickle
```

```python
plt.rcParams['figure.figsize'] = 40, 15
plt.rcParams['axes.labelsize']=92
plt.rcParams['xtick.labelsize']=64
plt.rcParams['ytick.labelsize']=64
plt.rcParams['legend.loc']='lower right'
plt.rcParams['legend.fontsize'] = 48
plt.rcParams['lines.linewidth'] = 4
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 1
plt.rcParams['grid.color'] = 'k'
plt.rcParams['boxplot.boxprops.linewidth'] = 6
plt.rcParams['boxplot.whiskerprops.linewidth']=6
plt.rcParams['boxplot.flierprops.markeredgewidth']=4
plt.rcParams['boxplot.capprops.linewidth'] =6
plt.rcParams['boxplot.medianprops.color']='C2'
plt.rcParams['boxplot.medianprops.linewidth'] =6
plt.rcParams['boxplot.meanprops.markersize'] = 20
```

```python
modes = ['ind', 'uni', 'fdiff', 'fmax', 'fsum', 'buc_eq', 'buc_con', 'buc_qeq', 'buc_qsd']
mode_names = ['ind', 'util', 'AEgal', 'MEgal', 'WUtil', 'buc_eq', 'buc_con', 'buc_qeq', 'buc_qsd']
modes2 = ['uni', 'fdiff', 'fmax', 'fsum', 'buc_eq', 'buc_con', 'buc_qeq', 'buc_qsd']
modes3 = ['ind', 'uni', 'fdiff', 'fmax', 'fsum', 'buc_con', 'buc_qeq', 'buc_qsd']
mode_names3 = ['ind', 'util', 'AEgal', 'MEgal', 'WUtil', 'buc_con', 'buc_qeq', 'buc_qsd']
```

```python
data_path = './data/age_k10_t100.pkl'
fo = open(data_path, 'rb')
data = pickle.load(fo)
fo.close()
```

```python
differences = data['differences']
```

```python
differences.boxplot(column = modes3, showmeans=True)
plt.xticks(range(1,len(modes3)+1), mode_names3)
plt.ylabel('Differences')
plt.tight_layout()
#plt.savefig('figs/race_total_errors.pdf')
plt.show()
```

```python
interferences = data['interferences']
interferences.boxplot(column = modes3, showmeans=True)
plt.xticks(range(1,len(modes3)+1), mode_names3)
plt.ylabel('Interferences')
plt.ylim((0,50))
plt.tight_layout()
#plt.savefig('figs/race_mean_ratio_errors.pdf')
plt.show()
```

```python
diff_ratios = data['diff_ratios']
diff_ratios.boxplot(column = modes3, showmeans=True)
plt.xticks(range(1,len(modes3)+1), mode_names3)
plt.ylabel('Ratio differences')
plt.tight_layout()
#plt.savefig('figs/race_mean_ratio_errors.pdf')
plt.show()
```

```python
inter_ratios = data['inter_ratios']
inter_ratios.boxplot(column = modes3, showmeans=True)
plt.xticks(range(1,len(modes3)+1), mode_names3)
plt.ylabel('Ratio Interferences')
plt.ylim((0,0.1))
plt.tight_layout()
#plt.savefig('figs/race_mean_ratio_errors.pdf')
plt.show()
```

```python
inter_maxs = data['inter_maxs']
inter_maxs.boxplot(column = modes3, showmeans=True)
plt.xticks(range(1,len(modes3)+1), mode_names3)
plt.ylabel('Max Interferences')
plt.yscale('log')
plt.tight_layout()
#plt.savefig('figs/race_mean_ratio_errors.pdf')
plt.show()
```

```python
inter_ratio_maxs = data['inter_ratio_maxs']
inter_ratio_maxs.boxplot(column = modes3, showmeans=True)
plt.xticks(range(1,len(modes3)+1), mode_names3)
plt.ylabel('Max Ratio Interferences')
plt.yscale('log')
plt.tight_layout()
#plt.savefig('figs/race_mean_ratio_errors.pdf')
plt.show()
```
